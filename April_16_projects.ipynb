```python
def check_str(data):
    valid_start = '({['
    valid_end = ']})'
    valid_paris = ['{}', '()', '[]']
    temp = []
    for i in data:
        if i in valid_start:
            temp.append(i)
        elif i in valid_end:
            if not temp:
                return False
            elif temp[-1] + i not in valid_paris:
                return False
            else:
                temp.pop(-1)
    if temp:
        return False
    return True
check_str("(){}[]")
```

```python
bank_name = 'HDFC Bank'
width = 40
bank_db = {}
max_retry = 3

err_1 = 'Error : Invalid input'
err_2 = 'ERROR : Maximum Retry Exceeded'
err_3 = 'ERROR : Invalid username or Password'

def print_header(mesg=None):
    print('-' * width)
    print(bank_name.center(width))
    if mesg:
        print(mesg.center(width))
    print('-' * width)
    
def welcome_screen(messages):
    print_header()
    valid_options = list(range(1,len(messages)+1))
    for i in range(max_retry):
        for sno,mesg in enumerate(messages,1):
            print(f'{sno}  {mesg}')
        ch = input('Enter your choice : ')
        if int(ch) in valid_options:
            return ch
        else:
            print(err_1)
    print(err_2)
    return False

def get_new_account_no():
    if bank_db:
        return max(bank_db) + 1
    else:
        return 1
    
def create_new_account():
    global bank_db
    print_header('NEW ACCOUNT OPENING')
    ano = get_new_account_no()
    name = input('Enter your name : ')
    paswd = input('Enter your Password : ')
    bank_db[ano] = {'name' : name,
                    'password' : paswd,
                    'balance' : 0
                   }
    print(f'Account Number {ano} created successfully')
    return True

def validate_user():
    print_header('Login Console')
    for i in range(max_retry):
        ano = int(input('Enter your Account No'))
        paswd = input('Enter your Password : ')
        if ano in bank_db and bank_db[ano]['password'] == paswd:
            return ano
        else:
            print(err_3)
    print(err_2)
    return False

def deposit(acc_no):
    global bank_db
    amt = int(input('Enter the Amount : '))
    bank_db[acc_no]['balance'] +=amt
    print(f'{amt} deposited successfully')

def withdraw(acc_no):
    global bank_db
    amt = int(input('Enter the Amount : '))
    if bank_db[acc_no]['balance'] >= amt:
        bank_db[acc_no]['balance'] -=amt
        print(f'{amt} Withdrawal success')
    else:
        print('ERROR : Insufficient Funds !!')

def check_balance(ano):
    name = bank_db[ano]['name']
    bal = bank_db[ano]['balance']
    print(f'{name} your Account balance is {bal}')
    
def logged_in_user_flow(ano):
    options = ['Deposit','Withdraw','check Balance','Logout']
    while True:
        ch = welcome_screen(options)
        if ch == '4':
            return
        elif ch == '1':
            deposit(ano)
        elif ch =='2':
            withdraw(ano)
        elif ch == '3':
            check_balance(ano)
```

```python
def main():
    while True:
        options = ['Create NEW Account','Login','Exit']
        choice = welcome_screen(options)
        if choice == '1':
            create_new_account()
        elif choice == '2':
            ano = validate_user()
            if ano:
                logged_in_user_flow(ano)
            else:
                print('ERROR : Login Unsuccessfull')
        elif choice == '3':
            print('Thanks for Banking with us')
            return 
```

```python
main()
```
